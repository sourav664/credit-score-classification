# Read Silver tables

```python
train_df = spark.table("credit_catalog.silver.train")
test_df = spark.table("credit_catalog.silver.test")
```

```python
train_df.display()
```

### Annual_Income and Monthly_Inhand_Salary are similar. Therefore, I am removing Annual_Income.

```python
cols_to_drop = ["Annual_Income"]

train_df = train_df.drop(*cols_to_drop)
test_df = test_df.drop(*cols_to_drop)
```

## Writing Gold Table

```python
train_df.write.mode("overwrite").format("delta").saveAsTable("credit_catalog.gold.ml_train")

test_df.write.mode("overwrite").format("delta").saveAsTable("credit_catalog.gold.ml_test")
```
