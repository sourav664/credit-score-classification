## Define Ingestion Configuration

```python
INGESTION_CONFIG = [
    {
        "path": "/Volumes/credit_catalog/credit_schema/raw_data/train.csv",
        "table": "train"
    },
    {
        "path": "/Volumes/credit_catalog/credit_schema/raw_data/test.csv",
        "table": "test"
    }
]
```

## Ingest Files into Bronze Tables

```python
for item in INGESTION_CONFIG:
    print(f"Ingesting into credit_catalog.bronze.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"credit_catalog.bronze.{item['table']}")
    )
```
